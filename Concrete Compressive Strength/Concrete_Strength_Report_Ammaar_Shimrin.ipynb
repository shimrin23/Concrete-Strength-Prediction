## Train the model in linear regression

### 1. Using fundamental equations


```python
# number of features
x_train_scaled.shape
```

```python
#initialize the weights
 
b_init =0.
w_init = np.array([ -0.  ,0.  ,0.  , 0.,0])
```

```python
#linear regression function
def predict(x, w, b): 
    p = np.dot(x, w) + b     
    return p 
```

```python
#cost function
def compute_cost(X, y, w, b): 
    m = X.shape[0]
    cost = 0.0
    for i in range(m):                                
        j_wb_i = np.dot(X[i], w) + b           #(n,)(n,) = scalar (see np.dot)
        cost = cost + (j_wb_i - y[i])**2       #scalar
    cost = cost / (2 * m)                      #scalar    
    return cost
```

```python
#This is to compute the gradient of the cost function
def compute_gradient(X, y, w, b): 
    m,n = X.shape           #(number of examples, number of features)
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w) + b) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w = copy.deepcopy(w_in)  #avoid modifying global w within function
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(X, y, w, b)   ##None

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               ##None
        b = b - alpha * dj_db               ##None
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```

```python
# Visualization Function
def plot_cost(J_history):
    plt.figure(figsize=(8, 5))
    plt.plot(range(len(J_history)), J_history, label="Cost Function")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.title("Cost Function vs. Iterations")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# initialize parameters
initial_w = np.zeros_like(w_init)
initial_b = 0.
# some gradient descent settings
iterations = 1500
alpha = 9.0e-3
# run gradient descent 
w_final, b_final, J_hist = gradient_descent(x_train_scaled, y_train, initial_w, initial_b,
                                                    compute_cost, compute_gradient, 
                                                    alpha, iterations)
plot_cost(J_hist)

print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = x_train.shape
for i in range(20):
    print(f"prediction: {np.dot(x_train_scaled[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
    
    
```

#### Model Evaluation for method 1


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you have the following arrays
# X_train, X_test, y_train, y_test

# Making predictions on the test set
y_train_pred = np.dot(x_train_scaled, w_final) + b_final
y_test_pred = np.dot(x_test_scaled, w_final) + b_final

# Calculating evaluation metrics
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
mae_train = mean_absolute_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"Training Set Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse_train:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_train:.2f}")
print(f"Mean Absolute Error (MAE): {mae_train:.2f}")
print(f"-> R-squared (R²): {r2_train:.2f}")

print(f"\nTest Set Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse_test:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_test:.2f}")
print(f"Mean Absolute Error (MAE): {mae_test:.2f}")
print(f"-> R-squared (R²): {r2_test:.2f}")
```

Scikit-learn's Linear Regression uses the **Normal Equation** to compute optimal weights directly. This non-iterative method is efficient for small-to-medium datasets, as it avoids iterations and hyperparameters. However, it can be computationally expensive for large datasets due to the complexity of matrix inversion in the equation.

## Additional Validation: Cross-Validation and SVR Hyperparameter Tuning

This section strengthens model validation using 5-fold cross-validation and GridSearchCV.

```python
from sklearn.model_selection import KFold, cross_validate, GridSearchCV
from sklearn.pipeline import Pipeline

# Reuse x, y, x_train, x_test, y_train, y_test from previous sections
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 1) Cross-validation for Linear Regression (with scaling inside pipeline)
lin_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

lin_cv = cross_validate(
    lin_pipe,
    x,
    y,
    cv=cv,
    scoring={'rmse': 'neg_root_mean_squared_error', 'r2': 'r2'}
)

print('Linear Regression - 5 Fold CV')
print(f"Mean RMSE: {-lin_cv['test_rmse'].mean():.3f} (+/- {lin_cv['test_rmse'].std():.3f})")
print(f"Mean R2  : {lin_cv['test_r2'].mean():.3f} (+/- {lin_cv['test_r2'].std():.3f})")

# 2) GridSearchCV for SVR (scaling inside pipeline)
svr_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVR())
])

param_grid = {
    'model__kernel': ['linear', 'rbf'],
    'model__C': [1, 10, 50, 100],
    'model__epsilon': [0.05, 0.1, 0.2],
    'model__gamma': ['scale', 'auto']
}

grid_svr = GridSearchCV(
    estimator=svr_pipe,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error',
    cv=cv,
    n_jobs=-1
)

grid_svr.fit(x_train, y_train)

print('\nBest SVR Parameters:')
print(grid_svr.best_params_)
print(f"Best CV RMSE: {-grid_svr.best_score_:.3f}")

best_svr = grid_svr.best_estimator_

y_pred_train_svr_tuned = best_svr.predict(x_train)
y_pred_test_svr_tuned = best_svr.predict(x_test)

mse_train_svr_tuned = mean_squared_error(y_train, y_pred_train_svr_tuned)
rmse_train_svr_tuned = np.sqrt(mse_train_svr_tuned)
mae_train_svr_tuned = mean_absolute_error(y_train, y_pred_train_svr_tuned)
r2_train_svr_tuned = r2_score(y_train, y_pred_train_svr_tuned)

mse_test_svr_tuned = mean_squared_error(y_test, y_pred_test_svr_tuned)
rmse_test_svr_tuned = np.sqrt(mse_test_svr_tuned)
mae_test_svr_tuned = mean_absolute_error(y_test, y_pred_test_svr_tuned)
r2_test_svr_tuned = r2_score(y_test, y_pred_test_svr_tuned)

print('\nTuned SVR - Training Metrics')
print(f"MSE : {mse_train_svr_tuned:.3f}")
print(f"RMSE: {rmse_train_svr_tuned:.3f}")
print(f"MAE : {mae_train_svr_tuned:.3f}")
print(f"R2  : {r2_train_svr_tuned:.3f}")

print('\nTuned SVR - Test Metrics')
print(f"MSE : {mse_test_svr_tuned:.3f}")
print(f"RMSE: {rmse_test_svr_tuned:.3f}")
print(f"MAE : {mae_test_svr_tuned:.3f}")
print(f"R2  : {r2_test_svr_tuned:.3f}")
```
